```python
import matplotlib.pyplot as plt 

pressure_variability= df.groupby('month')['pressure_mb'].std()

plt.figure(figsize = (10,6))
pressure_variability.plot(kind = 'bar', color = 'skyblue' , edgecolor = 'black')

plt.title('pressure variability(std dev) by month')
plt.xlabel('month')
plt.ylabel('standard deviation of pressure(hpa)')
plt.xticks(rotation = 45)
plt.show()
```

### in month 1 and 2 the bar are much longer which means the variability is higher and in the months between 3 to 12 the bars are much shorter and shows less variability .

```python
import matplotlib.pyplot as plt

x = df['air_quality_PM2.5']
y = df['temperature_celsius']
plt.figure(figsize=(8,6))
hb = plt.hexbin(x,y, gridsize =30 , cmap ='viridis',mincnt = 2)

plt.colorbar(hb , label = 'count of points')
plt.xlabel('airquality(g/m3)')
plt.ylabel('temperature(C)')
plt.title('temperature vs PM2.5 (hexbin plot)')
plt.show()
```

### here the hexbin plot tells us how many number of points lies to air qualtiy in relation to temperatures. vilet colour shows low points or low counts, only few areas (around 20'c to 30'c) have many repeated measurements the color is shown in green and yellowish.
